### dealing with missing value

```python
#### working on horse power column
```

```python
mean1=data['horsepower'].mean()
mean1
```

```python
#### beacuse horsepower is object data type
we need to convert it into float data
```

```python
data['horsepower']=data['horsepower'].astype(float)
```

```python
### error is could not convert string to float: '?' it is not helpfull
```

```python
data['horsepower']=data['horsepower'].replace('?',np.nan)
```

```python
median1=data['horsepower'].median()
median1
```

```python
#### now its working and succesfully convert it into object to float data
```

```python
data['horsepower'].replace(np.nan,median1,inplace=True)
```

```python
data.dtypes
```

```python
#### succesfully  deal with in object to float data
```
