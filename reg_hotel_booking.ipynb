```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

## Droping colums

```python
df.drop('company',axis=1,inplace=True)
```

```python
df.drop('hotel',axis=1,inplace=True)
```

```python
df.drop('meal',axis=1,inplace=True)
```

```python
df.drop('market_segment',axis=1,inplace=True)
```

```python
df.drop('agent',axis=1,inplace=True)
```

```python
df.drop('distribution_channel',axis=1,inplace=True)
```

## Replacing the missing value

```python
mean1=df['children'].mean()
mean1
```

```python
df['children'].replace(np.nan,mean1,inplace=True)
```

```python
mode1=df['country'].mode().values[0]
mode1
```

```python
df['country'].replace(np.nan,mode1,inplace=True)
```

```python
df.isnull().sum()
```

```python
duplicate=df.duplicated()
duplicate.sum()
```

### Encoding

```python
df=pd.get_dummies(df,columns=['arrival_date_month','country','reserved_room_type','assigned_room_type','deposit_type','customer_type','reservation_status_date','reservation_status'])
```

```python
df.dtypes
```

### Fitting model

```python
Y=df['is_canceled']
X=df.drop(["is_canceled"],axis=1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
model1=LogisticRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
model1.score(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```
